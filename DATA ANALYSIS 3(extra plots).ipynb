```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('output_parameters_with_phases.csv')
```

```python
df.columns = ['Ni', 'Al', 'Co', 'Fe', 'Cr', 'Δsₘᵢₓ', 'VEC', 'δ', 'Δχ', 'ΔHₘᵢₓ','Ω', 'Tm', 'no.of phases', 'phases', 'phase']
```

```python
df
```

```python
df= df.drop(['no.of phases','phases'], axis = 'columns')
```
